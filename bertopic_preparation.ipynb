```python
import matplotlib.pyplot as plt

def paint(df, column):
    fig, ax = plt.subplots(figsize=(18, 8))

    value_counts = df[column].value_counts()

    def cutting_labels(name):
        return name[:15]

    sns.barplot(x=value_counts.index.to_series().apply(cutting_labels), y=value_counts.values, ax=ax, palette="viridis")

    ax.set_xlabel('Channels')
    ax.set_ylabel('Number of recordings')
    ax.set_title('Distribution of channels')

    plt.xticks(rotation=45)

    return plt.show()
paint(df, column = 'channelname')
```

```python
def build_vocab(X):
    texts = X.apply(lambda x: x.split()).values
    vocab = {}

    for text in texts:
        for word in text:
            try:
                vocab[word] += 1
            except KeyError:
                vocab[word] = 1
    return vocab
```
